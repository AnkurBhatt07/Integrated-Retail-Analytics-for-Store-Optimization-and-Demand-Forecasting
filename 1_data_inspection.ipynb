```python
import numpy as np
# print numpy version 
print('np=',np.__version__)
import pandas as pd 
# print pandas version
print("pd=",pd.__version__)
```

```python
# reading all the datasets from the Retail Datasets folder
features = pd.read_csv('Retail Datasets/Features data set.csv')

sales = pd.read_csv('Retail Datasets/sales data-set.csv')

stores = pd.read_csv('Retail Datasets/stores data-set.csv')
```

```python
print("Features dataset shape:", features.shape)
print("Sales dataset shape:", sales.shape)
print("Stores dataset shape:", stores.shape)
print("\n\n")
print("features Dataset")
features.info()
print("\nMissing values in features dataset:\n", features.isnull().sum())
print("\n\n")
print("sales Dataset")
sales.info()
print("\nMissing values in sales dataset:\n", sales.isnull().sum())
print("\n\n")
print("stores Dataset")
stores.info()
print("\nMissing values in stores dataset:\n", stores.isnull().sum())
```

```python
# Date parsing and specifying dayfirst=True for correct date interpretation (day is first in the date format)
features['Date'] = pd.to_datetime(features['Date'],dayfirst=True)
sales['Date'] = pd.to_datetime(sales['Date'] , dayfirst=True)
```

```python
print("Unique Stores in features dataset:", features['Store'].nunique())
# Unique stores in sales dataset
print("Unique Stores in sales dataset:", sales['Store'].nunique())
# Unique stores in stores dataset
print("Unique Stores in stores dataset:", stores['Store'].nunique())
```

The missing values in the features dataset will be handled after merging all the three datasets. It is because after merging, we will have better contextual information to impute the missing values more accurately.



```python
# Merging datasets to create a comprehensive view
sales_features = pd.merge(sales, features, on=['Store', 'Date', 'IsHoliday'], how='left')
# sales is the core table.we want to retail all the sales records, so we do a left join.
final_df = pd.merge(sales_features,stores, on='Store', how='left')
```

```python
# final_df shape 
print("Final merged dataset shape:", final_df.shape)
# check if any rows got duplicated 
expected_rows = sales.shape[0]
actual_rows = final_df.shape[0]
print("Row count preserved after merge:", expected_rows == actual_rows)

# Check missing values in final_df 
final_df.isnull().sum().sort_values(ascending=False)
```

```python
# save the final_df for later reuse
# create the directory if it does not exist
import os
os.makedirs('processed_data', exist_ok=True)
 
final_df.to_csv('processed_data/final_merged_retail_data.csv', index=False)
```
